**Ejercicio 2: COLOREADO DE GRAFOS**

Los problemas de coloreado de grafos se originaron con el coloreado de mapas. Por ejemplo, considere cada estado de los Estados Unidos como un v√©rtice y conecte dos v√©rtices por un borde si y solo si comparten un l√≠mite, es decir, son vecinos. Si puede colorear el mapa de Estados Unidos usando k colores de tal manera que no haya dos estados vecinos que tengan el mismo color, entonces decimos que el mapa tiene una coloraci√≥n k. Mientras estudiaba en Londres a mediados del siglo XIX, el matem√°tico sudafricano Francis Guthrie conjetur√≥ a su profesor de matem√°ticas Augustus de Morgan que **cuatro colores son suficientes para colorear cualquier mapa**. Fue un problema abierto durante m√°s de 100 a√±os (solo probado por Appel y Haken en 1976)

El problema de coloreado de grafos **consiste en asignar colores a ciertos elementos de un grafo sujeto a ciertas restricciones.** 

El n√∫mero m√°s peque√±o de colores necesarios para colorear un grafo G se llama su **n√∫mero crom√°tico**. 



**> Coloreado de v√©rtices**

Dado un grafo no dirigido G = (V, E), una coloraci√≥n de v√©rtices es un etiquetado con colores de modo que no haya dos v√©rtices que compartan un borde que tengan el mismo color. Una coloraci√≥n que utiliza como m√°ximo k colores se denomina k-coloraci√≥n (propia).

El n√∫mero m√°s peque√±o de colores necesarios para colorear un grafo G se llama su n√∫mero crom√°tico (v√©rtice), y aqu√≠ se denota como **œáv(G)**. Un subconjunto de V asignado al mismo color se denomina clase de color. Un subconjunto S de V se llama conjunto independiente si no hay dos v√©rtices en S adyacentes en G. Por definici√≥n, cada clase de color forma un conjunto independiente.



```python
#EJEMPLO DE COLOREADO DE V√âRTICES DADO EL GRAFO G

d={}
vc = G.coloring()
#[
#[0, 5, 6, 3, 2, 11, 16], 
#[18, 19, 1, 7, 8, 17, 15], 
#[4, 12, 9], 
#[13, 10], 
#[14]
#]
colors = getListaColores(len(vc))
for c,l in zip(colors,vc):
    d[c]=l
G.show(vertex_colors = d, figsize = 15)
```

```python
#Funci√≥n que crea las cajas de los sudokus

def create_boxes(total_cells, box_width, boxes_row):
    boxes = []
    counter = 0
    for x in range(0,total_cells,box_width**3):
        for y in range(x,x+boxes_row*box_width,box_width):
            box=[]
            for i in range(y,y+box_width):
                for j in range(0,boxes_row):
                    box.append(i+j*boxes_row**2)
            boxes.append(box)
    return boxes
```

```python
#Funci√≥n que crea la matriz

def create_matrix(height,width,b_height,b_width,rows):
    no_vertices=height*width
    A=[]
    bxs=create_boxes(no_vertices,b_width,rows)
    for x in range(0,no_vertices):
        m=[0]*no_vertices
        column=x%height
        row=int(x/width)
        ###
        for y in range(row*width,(row*width)+width):
            if y != x:
                m[y] = 1
        ###
        for y in range(column,no_vertices,width):
            if y != x:
                m[y] = 1
        ###
        for b in bxs:
            if x in b:
                for n in b:
                    if n != x:
                        m[n] = 1
        A.append(m)
    return A
```

```python
#Clase para resolver el coloreo. Fue adaptada para que respetara un coloreo preasignado, es decir, las celdas que
#son utilizadas como pistas en el Sudoku y que determinan si se puede o no
#resolver

class Graph():
    
    def __init__(self,vertices):
        self.V = vertices
        self.graph=[[0 for column in range(vertices)]\
                    for row in range(vertices)]
        self.colour=[]
    
    def setC(self,colour):
        self.colour=colour
        
    def isSafe(self,v,colour,c):
        for i in range(self.V):
            if self.graph[v][i] == 1 and self.colour[i] == c and self.colour[i]!=0:
                return False
        return True
    
    def graphColourUtil(self,m,colour,v):
        if v==self.V:
            return True
        for c in range(1,m+1):
            if self.isSafe(v,colour,c)==True:
                if colour[v] == 0:
                    colour[v] = c
                elif self.graphColourUtil(m,colour,v+1)==True:
                    return True
                if self.graphColourUtil(m,colour,v+1)==True:
                    return True
                colour[v]=0
                
    def graphColouring(self,m):
        if self.graphColourUtil(m, self.colour,0)==False:
            return False
        
        counter=0
        for c in self.colour:
            print("vertice: "+str(counter+1) + "; color: "+str(c))
            counter+=1
        return True
```

```python
#Creaci√≥n de un grafo de 16 v√©rtices para un sudoku de 4x4

g = Graph(16)
no_vertices=16
no_colors=4

#Estado inicial

colors=[0]*no_vertices
colors[2]=4
colors[7]=2
colors[10]=1
colors[11]=3
colors[12]=3
colors[15]=4

#Creamos el objeto y le asignamos la matriz y el estado inicial

grafo=create_matrix(4,4,2,2,2)
g.graph=grafo
g.setC(colors)
g.graphColouring(4)
```
